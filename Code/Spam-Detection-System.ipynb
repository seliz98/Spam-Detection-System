```python
import json
import os

import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
data = {'reviewerID': [], 'reviewText': [], 'category': [], 'class': []}

with open('Cell_Phones_and_Accessories.json') as f:
    for line in tqdm(f):
        review = json.loads(line)
        data['reviewerID'].append(review['reviewerID'])
        data['reviewText'].append(review['reviewText'])
        data['category'].append(review['category'])
        data['class'].append(review['class'])
       
```

```python
df = pd.DataFrame(data)

print(df.shape)
df.head()
```

```python
df['reviewerID'] = df['reviewerID'].astype(str)
df['reviewText'] = df['reviewText'].astype(str)
df['category'] = df['category'].astype('category')
df['class'] = df['class'].astype(float)
```

```python
df['reviewText'] = df['reviewText'].str.lower()
```

```python
df.head(10)
```

```python
df.to_csv('amazon_reviews.csv', index=False)
```
